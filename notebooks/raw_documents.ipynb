```python
import pandas as pd

csv_files = sorted(glob.glob('../data/raw/*.csv'))
os.makedirs('../data/processed', exist_ok=True)

output_path = '../data/processed/ciciot2023_consolidado.parquet'
CHUNK_SIZE = 50000

first_write = True

for i, f in enumerate(csv_files):
    print(f"Procesando archivo {i+1}/{len(csv_files)}: {os.path.basename(f)}")
    
    for chunk in pd.read_csv(f, chunksize=CHUNK_SIZE):
        if first_write:
            chunk.to_parquet(output_path, index=False, engine='fastparquet')
            first_write = False
        else:
            chunk.to_parquet(output_path, index=False, engine='fastparquet', append=True)

print("Consolidacion completada")
```

```python
df = pd.read_parquet(output_path)
print(f"Total registros: {df.shape[0]:,}")
print(f"Total columnas: {df.shape[1]}")
print(f"\nDistribucion de clases:")
print(df['label'].value_counts())
```
